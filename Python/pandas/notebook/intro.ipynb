# Pandas

- Pandas is an open-source Python library used for data analysis and data manipulation.
- It provides powerful data structures such as Series and DataFrame.
- Pandas is widely used in Data Science, Machine Learning, and AI projects.
- It helps in cleaning, transforming, filtering, and analyzing data efficiently.
- Pandas can read and write data from various sources such as CSV, Excel, JSON, SQL databases, etc.
- It is built on top of NumPy and provides high-level data handling capabilities.

### Pandas Datastructure
- Series
- DataFrame
- Panel (rare use)
