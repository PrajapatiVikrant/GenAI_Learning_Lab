DataFrame
- A DataFrame is a two-dimensional labeled data structue in Pandas, similar to a table in a database, excel
spreadsheet, or a  SQL table.
- It consists of rows and columns, where:
  - Rows have indices (labels)
  - Columns have names (labels)

```python
# example
import pandas as pd 
data = {
    "calories":[420,380,390],
    "duration":[50,40,45]
}
print("Without dataframe:\n")
print(data)
print("\nWith DataFrame:\n")
print(pd.DataFrame(data))
```
