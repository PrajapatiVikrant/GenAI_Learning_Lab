# Properties of numpy array

```python
import numpy as np
arr = np.array([1,2,3,4,5,6])
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
```

1. shapes:- used to find the size of matrix 

# 

```python

print(f"Shape of arr: {arr.shape}")
print(f"Shape of arr1: {arr1.shape}")
print(f"Shape of arr2: {arr2.shape}")
```

2. ndim :- it helps to find dimension of any array

```python
print(f"Dimensions of arr: {arr.ndim}")
print(f"Dimensions of arr1: {arr1.ndim}")
print(f"Dimensions of arr2: {arr2.ndim}")
```
