```python
import numpy as np
arr = np.array([1,2,3,4,5])
```

```python
# sum
print(np.sum(arr))   #  summ of all array's elements
print(np.mean(arr))  # average of all array's elements
print(np.min(arr))   # find minimum element of array
print(np.max(arr))   # find maximum element of array
print(np.std(arr))   # calculate standard deviation
print(np.var(arr))   # calculate variant
```
