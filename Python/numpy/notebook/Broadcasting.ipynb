# Broadcasting

#### Problem

##### If we want to perform specing operation or calculation with each element of array like a array contain many element of number and you want to calculate all element's  10% 

Without BroadCasting

```python
import numpy as np
arr = [100,300,240,980]
discount = 10;
result = []
for i in arr:
    result.append(i-(i*discount/100))

print(result)

"""
This is a true method but when list contain many items then loop executes very slow

"""
```

With Broadcasting

```python
arr = np.array([100,300,240,980])
discount = 10;

result = arr - (arr*discount/100)
print(result)
'''
here we can perform operation directly without element
'''
```
