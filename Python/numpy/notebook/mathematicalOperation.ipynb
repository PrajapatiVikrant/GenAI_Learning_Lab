```python
import numpy as np
```

Mathamatical Operation in 1-D ARRAY

```python
arr = np.array([1,2,3,4,5])
```

```python
arr = arr+2
print(arr)
```

```python
arr = arr*2
print(arr)
```

```python
arr = arr-2
print(arr)
```

```python
arr = arr/2
print(arr)
```

```python
arr = arr % 2
print(arr)
```

Mathematical Operation in 2-D ARRAY

```python
arr = np.arange(9).reshape(3,3)
print(arr)
```

```python
arr = arr+2
print(arr)
```

```python
arr = arr-1
print(arr)
```

```python
arr = arr*3
print(arr)
```

```python
arr = arr%2
print(arr)
```

Mathematical Operation between two matrices

```python
arr2 = np.arange(9).reshape(3,3)
print(arr2)
```

```python
addition = arr+arr2
print(addition)
```

```python
mul = addition * arr2
print(mul)
```
