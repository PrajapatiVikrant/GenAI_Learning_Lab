Creating array from python list

```python
import numpy as np

arr = np.array([1,2,3,4,5,6])
print(arr)
```

Default Values

```python
arr = np.zeros(6)
print(arr)
```

Ones function 

```python
arr = np.ones(6)
print(arr)
```

Full function to creating array where every element start with a specific number
np.full(shapes,number)

```python
arr = np.full(6, 42)
print(arr)
```

Creating sequences of numbers in numpy

arange()

```python
arr = np.arange(5)   # Start at 0, end before 5
print(arr)
```

```python
arr = np.arange(1,5)   # Start at 1, end before 5 and can be start array from any number like 1 to 5, 2 to 10, 100 to 200 etc.
print(arr)
```

```python
arr = np.arange(1, 10, 2) # Start at 1, end before 10 and step by 2
print(arr)
```

 Creating identity matrices

```python
arr = np.eye(3)  # Identity matrix of size 3x3 with 1s on the diagonal and 0s elsewhere
print(arr)
```
